Calculus - Branch of mathematics that studies continuous change

input -> function -> output
Domain - all valid inputs a function can take
Range - all possible outputs values the function can produce

Differentiation describes how fast one quantity changes compared to another.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("Differentiation.png")

plt.figure(figsize=(10, 14))
plt.imshow(img)
plt.axis("off")
plt.show()
```
