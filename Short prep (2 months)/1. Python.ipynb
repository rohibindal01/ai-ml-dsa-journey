## 6. Operators
Used between operands. In `a + b`, `a` and `b` are operands and `+` is the operator.

### 6a. Arithmetic Operators

````python
print(3 + 5)   # Addition
print(3 - 5)   # Subtraction
print(3 * 5)   # Multiplication
print(3 / 5)   # Division (returns float)
print(3 % 5)   # Modulus (remainder)
print(3 // 5)  # Floor division (rounds down)
print(3 ** 5)  # Exponentiation (3 raised to the power 5)
````

### 6b. Comparison Operators
These return `True` or `False`.

````python
print(3 > 5)   # Greater than
print(3 < 5)   # Less than
print(3 >= 5)  # Greater than or equal to
print(3 <= 5)  # Less than or equal to
print(3 == 5)  # Equal to — checks if values are equal (NOT assignment)
print(3 != 5)  # Not equal to
````

### 6c. Assignment Operators
`=` assigns a value. `+=`, `-=`, etc. are shorthand update operators.

````python
a = 9      # assigns 9 to a
a += 2     # same as a = a + 2
print(a)   # 11
````

### 6d. Logical Operators

| Operator | Description |
|----------|-------------|
| `and` | True only if **both** are True |
| `or` | True if **at least one** is True |
| `not` | Flips the boolean |

````python
print((3 > 5) and (3 < 5))   # False
print((3 > 5) or (3 < 5))    # True
identifier = False
print(not identifier)         # True
````

## 10. Loops
### While Loop

````python
count = 1
while count <= 10:
    print('hello', count)
    count += 1
````

**`break`** — exits the loop immediately  
**`continue`** — skips the current iteration

````python
# break example
i = 1
while True:
    if i == 4:
        break
    print(i)
    i += 1
````

````python
# continue example
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
````

### For Loop & range()
`range(start, stop, step)`
- `start` — where the sequence begins (included, default = 0)
- `stop` — where it ends (**not** included)
- `step` — increment (default = 1)

````python
for i in range(0, 5, 1):
    print(i)
print('----------------------------------')
for i in range(5):
    print(i)
print('----------------------------------')
for i in range(0, 5, 2):
    print(i)
````

## 14. Object-Oriented Programming (OOP)
- **Class** — Blueprint of an object (no memory allocated)
- **Object** — Instance of a class (memory allocated)

````python
class Hello:
    name = 'Rohi'
    age = 25
    college = 'Stevens'

object1 = Hello()
print(object1.name, object1.age, object1.college)
````

### `__init__` Method
Called automatically every time a new object is created.

````python
class HelloWithInit:
    college = 'Stevens'   # class attribute — shared by all instances

    def __init__(self, name, age):
        self.name = name   # instance attribute — unique per object
        self.age = age

object2 = HelloWithInit('Rahul', 80)
print(object2.name, object2.age, object2.college)
````

### Instance, Class & Static Methods
| Method type | Decorator | Can access |
|-------------|-----------|------------|
| Instance method | *(none)* | class + instance attributes |
| Class method | `@classmethod` | class attributes only |
| Static method | `@staticmethod` | neither |

````python
class Elephant:
    teeth = 2

    def __init__(self, legs, trunks):
        self.legs = legs
        self.trunks = trunks

    def wohoo(self):           # instance method
        print(f'Elephant has {self.legs} legs, {self.trunks} trunks and {self.teeth} teeth')

    @classmethod
    def voo(cls):              # class method
        print(f'Elephant has {cls.teeth} teeth')

    @staticmethod
    def hey(friends, enemies): # static method
        print(f'The elephant has {friends} friends and {enemies} enemies')

elephant1 = Elephant(4, 1)
elephant1.wohoo()
elephant1.voo()
elephant1.hey(6, 8)
````

### 1. Encapsulation
Wrapping data and methods into a single unit (class) and controlling access.

| Access level | Syntax | Who can access |
|---|---|---|
| Public | `self.name` | Anyone |
| Protected | `self._name` | Class + subclass (by convention) |
| Private | `self.__name` | Only within the class (name-mangled) |

````python
class PersonName:
    def __init__(self, name, namee, nameee):
        self.name = name           # public
        self._namee = namee        # protected
        self.__nameee = nameee     # private

name1 = PersonName('Rohi', 'Riya', 'Tina')

print(name1.name)                    # public — works directly
print(name1._namee)                  # protected — accessible by convention
print(name1._PersonName__nameee)     # private — requires name mangling syntax
````

### 2. Abstraction
Hides implementation details and shows only essential functionality.  
Abstract classes force child classes to implement all abstract methods.

````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print('Car is starting')

c = Car()
c.start()
````

### 3. Inheritance
Reusing attributes and methods from a parent class.

| Type | Description |
|------|-------------|
| Single | Child inherits from one parent |
| Multilevel | Chain: A → B → C |
| Multiple | Child inherits from more than one parent |
| Hybrid | Combination of the above |

````python
class Parent:
    attribute1 = 'HI'
    attribute2 = 'HELLO'

    def method(self, attribute3):
        self.attribute3 = attribute3
        return attribute3

class Child(Parent):
    def __init__(self, attribute4):
        self.attribute4 = attribute4

child1 = Child('HEYA')
print(child1.attribute1, child1.attribute2, child1.attribute4, child1.method('HEY'))
````

### 4. Polymorphism
The same method name behaves differently depending on the object.

- **Method Overriding** — child class provides its own version of a parent method
- **Duck Typing** — Python checks if the object *has* the method, not what type it is

````python
class Dog:
    def speak(self):
        print('Bark')

class Cat:
    def speak(self):
        print('Meow')

def make_sound(animal):
    animal.speak()   # works as long as the object has a speak() method

make_sound(Dog())   # Bark
make_sound(Cat())   # Meow
````

## 17. Exception Handling
| Block | Purpose |
|-------|---------|
| `try` | Code that might raise an error |
| `except` | Handles the specific error |
| `else` | Runs only if no error occurred |
| `finally` | Always runs, regardless of error |

````python
try:
    x = int(input('Enter a number: '))
    ans = 10 / x
except ZeroDivisionError:
    print('Dividing by 0 is not allowed!')
else:
    print(f'Answer = {ans}')
finally:
    print('This always gets printed.')
````
