Database: It is a collection of data in a format that can be easily accessed.
Why Databases?
1. can store large data
2. easier to insert, update or delete data

SQL- Relational Database (data stored in tables). eg. MYSQL, Oracle, etc
NoSQL- Non Relational Database (data stored in document/ key-value/ graphs etc). eg. MongoDB, cASSANDRA 

Structured Query Language (SQL)
SQL is a programming language used to interact with relational databases. (table 1 and table 2 are related)

CREATE DATABASE nameofthetable; #create the table
DROP DATABASE nameofthetable; #delete the table
SHOW DATABASE;
SHOW TABLES;

#Structure of the table
CREATE DATABASE nameofthetable(
    columnname1 datatype constraint ,
    columnname2 datatype constraint
);

Constraints: Rules for data in the table
1. NOT NULL - Columns cannot have a null value
2. UNIQUE - All values in the column are different
3. DEFAULT - Sets the default value of the column
4. CHECK - It can limit the values allowed in a column

Primary Key: makes a column unique and not null but used only for one.

CREATE TABLE temp(
    id int not null,
    PRIMARY KEY (id)
);

Foreign Key: column or a set of columns
- Creates relationship between tables
- References Primary Key of another table
- There can be multiple and duplicate FKs
Syntax:

FOREIGN KEY (column_name)
REFERENCES ParentTable(primary_key)

Example:

FOREIGN KEY (Students_StudentID) REFERENCES Students(StudentID)

OPERATORS
Arithmetic operators: +,-,*,/,%
Comparison operators: =,!=,>,<,>=,<=
Logical operators: AND, OR, NOT, IN, BETWEEN, ALL, LIKE, ANY
Bitwise operators: bitwise and (&), bitwise or (|)

Limit Clause: Sets an upper limit on the number of the rows to be returned
SELECT col1, col2 FROM table_name
LIMIT number;

ORDER BY: to sort in ascending and descending order
SELECT col1,col2 FROM table_name
ORDER BY col_name ASC;

Aggreagate functions: They perform a calculation on the set of values, and return a single value.
COUNT()
MIN()
MAX()
SUM()
AVG()

GROUP BY: groups rows that have same value into summary rows. 
SELECT col1, col2
FROM tablename
GROUP BY col_name;

HAVING: similar to where i.e. applies some conditions on the rows. It is used when we want to apply any condition after grouping
SELECT col1, col2
FROM tablename
GROUP BY col_name
HAVING condition;


General Order
SELECT columns FROM tablename WHERE condition GROUP BY column HAVING condition ORDER BY column ASC;

UPDATE: to update the existing rows
UPDATE tablename
SET col1 = val1
WHERE condition;

DELETE: to delete existing rows
DELETE FROM tablename
WHERE condition;

ALTER: to change the schema

add column
ALTER TABLE tablename
ADD COLUMN columnname datatype constraint;

drop column
ALTER TABLE tablename
DROP COLUMN columnname;

rename table
ALTER TABLE tablename
RENAME TO newname;

Truncate: To delete table's data
TRUNCATE TABLE tablename;

Transactions
ACID Properties
ATOMICITY: all statements succeed or none succeed
CONSISTENCY: data moves from one valid state to another
ISOLATION: parallel transactions dont interfere
DURABILITY: committed data is permanently saved

Transactions
disable autocommit: SET autocommit = 0
enable autocommit: SET autocommit = 1

START TRANSACTION;
UPDATE ...............
SAVEPOINT after_wallet;
UPDATE ...............
ROLLBACK TO after_wallet; #this undoes everything after the savepoint
COMMIT;

similarly for LEFT JOIN, RIGHT JOIN, OUTER JOIN, SELF JOIN(Table joins with itself) and CROSS JOIN(Combine every item from the first table with every item from the second table.)

Subqueries
Query within another query. It involves 2 select statements.

SELECT columns
FROM tablename
WHERE colnameoperator
(subquery);

Views
It is a virtual table based on the result set of an sql statement

CREATE VIEW view1 AS
SELECT col1,col2 FROM tablename;

Index
These are special database objects that make data retrieval faster

CREATE INDEX idxname ON tablename(columnname);
DROP INDEX idxname ON tablename;

Composite Index (index created on more than one column)
CREATE INDEX index_name
ON table_name(column1, column2);

Stored Procedures: Predefined set of SQL statements that you can save in the database and execute whenever needed. Instead of writing the same SQL queries again and again, you save them once and call them whenever needed.

Create:

CREATE PROCEDURE procedure_name()
AS
BEGIN
    SQL statements
END;

Execute:

EXEC/CALL procedure_name();

Delete:

DROP PROCEDURE procedure_name();
