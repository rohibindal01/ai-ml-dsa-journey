Showing the relationships between different kinds of data.

```python
#follow this syntax for every kind of plot
import matplotlib.pyplot as plt
X = [1,2,3,4]
Y = [5,6,7,8]
plt.plot(X,Y) #only this will change as per the need
plt.title('example')
plt.xlabel('X')
plt.ylabel('Y')
# plt.legend()- A legend is a small box on a graph that explains what each color, line, or symbol represents.
plt.show
```

```python
#plot methods
plt.plot(x, y) - Line plot
plt.bar(x, height) - Vertical bar chart
plt.barh(y, width) - Horizontal bar chart
plt.scatter(x, y) - Scatter plot
plt.pie(data, labels=) - Pie chart
plt.hist(data, bins) - Histogram
plt.boxplot(data) - Box plot
#lower = Q1-(1.5*IQR)
#upper = Q1+(1.5*IQR)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("boxplot.png")

plt.figure(figsize=(10, 14))
plt.imshow(img)
plt.axis("off")
plt.show()
```
