Probability is a measure of the likelihood that a particular event will occur (ranges from 0 to 1)

sample space (S)- all possible outcomes
event (E)- a subset of outcomes
P(E)- probability of event E

P(E)= n(E)/n(S)

P(A∩B)= Probability that A and B both occur
P(AUB)= Probability that A or B both occur

1. Mutually Exclusive Events- Two events that cannot happen at the same time
2. Independent Events- The outcome of one event does not affect the probability of the other
3. Dependent Events- The outcome of one event affects the probability of the other
4. Complementary Events- It is if one occurs exactly when the other does not  

Rules of Probability
1. Complementary Rule- If A is an event then its complement A^c has probability

P(A^c)= 1-P(A)

2. Addition Rule

Mutually exclusive events- If A and B cannot happen together
P(A or B)=P(AUB)= P(A)+P(B)

Non Mutually exclusive events- If A and B can happen together
P(AUB)= P(A)+ P(B)-P(A∩B)

3, Multiplication Rule

Independent Events- If A and B are independent
P(A and B)= P(A∩B) = P(A) X P(B)

Dependent events- If A and B are dependent
P(A and B)= P(A∩B) = P(A)XP(B|A)

Conditional Probability

P(B|A)= P(A∩B) | P(A)

Law of total probability
It is used to find the probability of an event B by breaking it ionto several mutually exclusive cases.

P(B)= P(B|A1)*P(A1)+P(B|A2)*P(A2)+.........+P(B|An)*P(n)

Where: 
A1,A2,...,AN are mutually exclusive events (one of tehm must happen)

P(A\B)- Probability of event A given B has occured

Random Variables- It is a numerical quantity that depends on the outcome of a random experiment

Discrete RV- countable values
Continuous RV- any value in a range(fractions)

mean- average value
median- middle value
mode- most frequent value
weighted mean- some values contribute more than the other values
Variance tells you how spread out the data is from the average (mean).
Standard deviation also measures how spread out the data is, but it's easier to understand because it's in the same units as the data.

Probability Distribution- It describes how probabilities are assigned to all the possible outcomes. It tells you how likely different outcomes are.

Discrete probability distribution- outcomes we can count. eg. Binomial distribution

Continuous probability distribution- outcomes can be any number in range. eg. uniform and normal distribution

Binomial distribution- used to model the number of successes in N trials

Uniform distribution- where all outcomes aare equally likely

Normal distribution- bell shaped and symmetric around the mean. most values cluster around the mean.  value sfar from the mean are less likely.

Central limit theorem
