Selecting data

1. Selecting columns
df[["Name", "Marks"]]

2. Selecting Rows (loc and iloc)
df.loc[0:2] #returns rows 0,1 and 2
df.iloc[0:2] #returns rows 0 and 1

3. Selecting rows and columns
df.loc[0:2, ["Name", "Marks"]] #df.loc[row_label, column_name]
df.iloc[0:3, 0:2] #df.iloc[row_positions, column_positions]

4. Selecting single scalar value (at and iat)
df.at[1, "Marks"] #df.at[row_label, column_name]
df.iat[1, 2] #df.iat[row_position, column_position]

Transforming Data (Feature Engineering)

1. apply()
   Apply a function to each value

   Example:
   df["Marks"].apply(lambda x: x*2)

2. map()
   Replace or transform values in a column

   Example:
   df["Gender"].map({"M":"Male","F":"Female"})

3. assign()
   Create new columns

   Example:
   df.assign(Bonus=df["Marks"]+5)

4. replace(old,new)
   Replace values

   Example:
   df["Gender"].replace("Male","M")

5. rename()
   Rename columns

   df.rename(columns={"old":"new"})

6. sort_values()
   Sort by column values

   df.sort_values("Marks")

7. sort_index()
   Sort by index

   df.sort_index()

8. reset_index()
   Reset row numbering

   df.reset_index(drop=True)

9. rank()
   Assign rankings

   df["Rank"] = df["Marks"].rank()

10. Reorder Columns

   df = df[["Name","Marks","Age"]]

Grouping & Aggregation

melt()  → Columns become Rows
   Converts Wide Format → Long Format

   Syntax:
   pd.melt(
       df,
       id_vars,
       value_vars,
       var_name,
       value_name
   )

   Example:

   Original DataFrame

   Student  Math  Science
   Rohi      90      80
   Aman      85      95

   pd.melt(
       df,
       id_vars=["Student"],
       value_vars=["Math","Science"],
       var_name="Subject",
       value_name="Marks"
   )

   Output

   Student  Subject   Marks
   Rohi      Math      90
   Aman      Math      85
   Rohi    Science     80
   Aman    Science     95

   Parameters:
   id_vars    → Column(s) to keep
   value_vars → Columns to convert into rows
   var_name   → Name of new column storing old column names
   value_name → Name of new column storing values


pivot() → Rows become Columns
   Converts Long Format → Wide Format

   Syntax:
   df.pivot(
       index,
       columns,
       values
   )

   Example:

   Original DataFrame

   Student  Subject   Marks
   Rohi      Math      90
   Rohi    Science     80
   Aman      Math      85
   Aman    Science     95

   df.pivot(
       index="Student",
       columns="Subject",
       values="Marks"
   )

   Output

   Subject  Math  Science
   Student
   Aman      85      95
   Rohi      90      80

   Parameters:
   index   → Column to keep as rows
   columns → Column whose values become new column names
   values  → Values to fill in the table

Merging & Joining Data

Syntax:

pd.merge(
    df1,
    df2,
    on="ID",
    how="join_type"
)

left_df  → First DataFrame

right_df → Second DataFrame

on       → Common column

how      → inner / left / right / outer

Join Types:

1. Inner Join
   Keep matching rows only

   pd.merge(df1, df2,
            on="ID",
            how="inner")

2. Left Join
   Keep ALL rows from LEFT DataFrame and Add matching rows from RIGHT DataFrame

   pd.merge(df1, df2,
            on="ID",
            how="left")

3. Right Join
   Keep ALL rows from RIGHT DataFrame and Add matching rows from LEFT DataFrame

   pd.merge(df1, df2,
            on="ID",
            how="right")

4. Outer Join
   Keep all rows from both DataFrames

   pd.merge(df1, df2,
            on="ID",
            how="outer")

Data Concatenation: Used to combine multiple DataFrames.

Syntax:

pd.concat([df1, df2])

1. Row-wise Concatenation
   (Vertical Stacking)

   pd.concat([df1, df2])

   Output:
   df1
   ↓
   df2

2. Column-wise Concatenation
   (Horizontal Stacking)

   pd.concat(
       [df1, df2],
       axis=1
   )

axis Values:

axis=0 → Rows (Default)
axis=1 → Columns
