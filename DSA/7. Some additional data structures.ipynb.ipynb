1. Counter returns the frequency of each item in the iterable
Counter(iterable)

```python
from collections import Counter
votes = ['Rohi', 'Huma', 'Rohi', 'Rita', 'Huma', 'Rohi']
result = Counter(votes)
print(result)
```

2. ChainMap groups multiple dictionaries together and treats them as a single dictionary.
ChainMap(dictionary1(preferred), dictionary2)

```python
from collections import ChainMap
defaults = {"theme": "light" , "language": "English", "font_size": 12}
user_settings = {"theme": "dark", "font_size": 14}
config = ChainMap(user_settings, defaults)
print(config["theme"])
print(config["language"])
print(config["font_size"])
```

3. Defaultdict: Unlike a normal dictionary, if you access a key that doesn't exist, it automatically creates a default value instead of raising a KeyError.
defaultdict(defaultvalue)

```python
# Import defaultdict from the collections module
from collections import defaultdict

# Create a defaultdict where the default value is an empty list
students_by_grade = defaultdict(list)

# Add "rohi"(value) to the list of Grade A students(a key)
students_by_grade["A"].append("rohi")

# Add "huma"(value) to the list of Grade B students(a key too)
students_by_grade["B"].append("huma")

# Add another student to Grade A
students_by_grade["A"].append("riya")

# Print students of Grade C
# Since Grade C doesn't exist, defaultdict automatically creates an empty list for it.
print(students_by_grade["C"])
print(students_by_grade['B'])
print(students_by_grade['A'])
```

4. Dequeue: Insertion and deletion can be performed from both the front and the rear.

```python
# Import deque from the collections module
from collections import deque

# Create an empty deque
queue = deque()

# Add 'rohi' to the right (end) of the deque
queue.append('rohi')

# Add 'huma' to the right (end) of the deque
queue.append('huma')

# Add 'vip' to the left (front) of the deque
queue.appendleft('vip')

# Remove an element from the right (end)
queue.pop()

# Remove an element from the left (front)
queue.popleft()

# Print the remaining elements
print(queue)
```

5. Namedtuple: It lets you give names (labels) to the elements of a tuple, making your code easier to read and understand.

```python
# Import namedtuple from the collections module
from collections import namedtuple

# Create a new data type called Student
# It has three fields:
# 1. name
# 2. grade
# 3. score
Student = namedtuple("Student", ["name", "grade", "score"])

# Create a Student object
s1 = Student("huma", "A", 95)

# Access the student's name
print(s1.name)

# Access the student's grade
print(s1.grade)

# Access the student's score
print(s1.score)

# Access the first value using its index
print(s1[0])
```

6. OrderedDict: An OrderedDict is a special dictionary from Python's collections module that remembers the order in which items are inserted.

```python
# Import OrderedDict from the collections module
from collections import OrderedDict

# Create an empty OrderedDict
cart = OrderedDict()

# Add items to the cart
cart["apple"] = 2
cart["banana"] = 5
cart["mango"] = 1

# Move "apple" to the end of the dictionary
cart.move_to_end("apple")

# Traverse the OrderedDict and print each item with its quantity
for item, qty in cart.items():
    print(item, qty)
```
