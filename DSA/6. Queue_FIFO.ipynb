Functions

```python
class Queue:

    # Constructor
    def __init__(self):
        # Create an empty queue
        self.queue = []

    # Add an element at the rear of the queue
    def enqueue(self, data):
        self.queue.append(data)

    # Remove an element from the front of the queue
    def dequeue(self):

        # Check if queue is empty
        if len(self.queue) == 0:
            return "Queue Empty"

        # Remove and return the first element
        return self.queue.pop(0)

    # Display the front element without removing it
    def peek(self):

        # Check if queue is empty
        if len(self.queue) == 0:
            return "Queue Empty"

        # Return the first element
        return self.queue[0]


# Create Queue object
q = Queue()

# Add elements
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

# Display front element
print(q.peek())

# Remove front element
print(q.dequeue())

# Display new front element
print(q.peek())
```

Linkedlist in queue

```python
# Node class represents a single node in the linked list
class Node:

    # Constructor
    def __init__(self, data):

        # Store the value of the node
        self.data = data

        # Initially, the node does not point to any other node
        self.next = None


# Queue class
class Queue:

    # Constructor
    def __init__(self):

        # Front points to the first node of the queue
        self.front = None

        # Rear points to the last node of the queue
        self.rear = None


    # Insert an element at the rear of the queue
    def enqueue(self, data):

        # Create a new node
        new_node = Node(data)

        # If the queue is empty
        if self.front is None:

            # Both front and rear should point to the new node
            self.front = new_node
            self.rear = new_node

        # Otherwise, queue already has elements
        else:

            # Connect the current rear node to the new node
            self.rear.next = new_node

            # Move the rear pointer to the new node
            self.rear = new_node


    # Remove an element from the front
    def dequeue(self):

        # If the queue is empty
        if self.front is None:
            print("Queue is Empty")
            return

        # Store the front value before removing it
        removed = self.front.data

        # Move front to the next node
        self.front = self.front.next

        # If queue becomes empty after deletion
        if self.front is None:

            # Rear should also become None
            self.rear = None

        # Return the removed value
        return removed


    # Display all queue elements
    def display(self):

        # If queue is empty
        if self.front is None:
            print("Queue is Empty")
            return

        # Start from the front node
        temp = self.front

        # Traverse until temp becomes None
        while temp:

            # Print current node data
            print(temp.data, end=" -> ")

            # Move to next node
            temp = temp.next

        # End of linked list
        print("None")


# Create Queue object
q = Queue()

# Insert elements
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)

# Display queue
print("Queue:")
q.display()

# Remove first element
print("Removed:", q.dequeue())

# Display queue again
print("Queue After Dequeue:")
q.display()
```

Using stack in queue

```python
# Queue implementation using two stacks
class Queue:

    # Constructor
    def __init__(self):

        # Stack 1 is used for enqueue operations
        self.stack1 = []

        # Stack 2 is used for dequeue operations
        self.stack2 = []


    # Add an element to the queue
    def enqueue(self, data):

        # Push the element onto stack1
        self.stack1.append(data)


    # Remove an element from the queue
    def dequeue(self):

        # If both stacks are empty, then the queue is empty
        if not self.stack1 and not self.stack2:
            return "Queue is Empty"

        # If stack2 is empty,
        # transfer all elements from stack1 to stack2
        if not self.stack2:

            # Move every element from stack1
            # to stack2 one by one
            while self.stack1:

                # Pop from stack1
                # and push into stack2
                self.stack2.append(self.stack1.pop())

        # Remove the top element from stack2
        # This is the oldest element of the queue
        return self.stack2.pop()


# Create Queue object
q = Queue()

# Insert elements
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

# Remove elements
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
```
