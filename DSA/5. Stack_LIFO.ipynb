Some basic functions

```python
stack = []

# Push
stack.append(10)
stack.append(20)
stack.append(30)

print("Stack:", stack)

# Peek
print("Top Element:", stack[-1])

# Pop
removed = stack.pop()
print("Removed:", removed)

print("Final Stack:", stack)
```

Example 

```python
# Function to check whether brackets are balanced
def check_brackets(expr):

    # Create an empty stack to store opening brackets
    stack = []

    # Loop through each character in the input string
    for ch in expr:

        # If the character is an opening bracket
        if ch in "([{":

            # Push (store) the opening bracket onto the stack
            stack.append(ch)

        # If the character is a closing bracket
        elif ch in ")]}":

            # If the stack is empty, there is no opening bracket to match this closing bracket
            if not stack:
                return False

            # Remove the top (most recent) opening bracket
            top = stack.pop()

            # Check whether the popped opening bracket
            # matches the current closing bracket

            # Example:
            # ')' should match '('
            # '}' should match '{'
            # ']' should match '['

            if (ch == ')' and top != '(') or \
               (ch == '}' and top != '{') or \
               (ch == ']' and top != '['):

                # Brackets don't match
                return False

    # After checking all characters,
    # if the stack is empty, every opening bracket
    # had a matching closing bracket.
    # Otherwise, some opening brackets were never closed.
    return len(stack) == 0


# Test the function
print(check_brackets("{[()]}"))
```
