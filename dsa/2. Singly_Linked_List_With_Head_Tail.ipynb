```python
# Node class represents a single node in the linked list
class Node:
    def __init__(self, data):
        # Store the value/data of the node
        self.data = data

        # Initially, the node does not point to any other node
        self.next = None
```

```python
# Singly Linked List class
class SinglyLinkedList:
    def __init__(self):
        # Head points to the first node in the list
        self.head = None

        # Tail points to the last node in the list
        self.tail = None


    # Function to add a new node at the end of the list
    def insert_end(self, data):

        # Create a new node with the given data
        new_node = Node(data)

        # Check if the list is empty
        if self.head is None:

            # If empty, the new node becomes the first node (head)
            self.head = new_node

            # Since there is only one node, it is also the tail
            self.tail = new_node

            # Exit the function
            return

        # If the list is not empty:

        # Connect the current tail node to the new node
        self.tail.next = new_node

        # Move the tail pointer to the new last node
        self.tail = new_node


    # Function to count the number of nodes in the list
    def size(self):

        # Initialize count to 0
        count = 0

        # Start traversing from the head node
        current = self.head

        # Continue until we reach the end (None)
        while current is not None:

            # Count the current node
            count += 1

            # Move to the next node
            current = current.next

        # Return total number of nodes
        return count
    
    # Function to display all elements of the linked list
    def display(self):

        # Start from the first node (head)
        current = self.head

        # Traverse until we reach the end of the list
        while current is not None:

            # Print the data of the current node
            # end=" -> " prevents moving to a new line
            print(current.data, end=" -> ")

            # Move to the next node
            current = current.next

        # After the last node, print None to indicate end of list
        print("None")

    # Function to delete the first occurrence of a value
    def delete(self, value):

        # Check if the list is empty
        if self.head is None:
            print("List is empty")
            return

        # If the value is in the first node (head)
        if self.head.data == value:

            # Move head to the next node
            self.head = self.head.next

            # If list becomes empty, update tail also
            if self.head is None:
                self.tail = None

            return

        # Start traversal from head
        current = self.head

        # Traverse until the end
        while current.next is not None:

            # Check if the next node contains the value
            if current.next.data == value:

                # If deleting the tail node
                if current.next == self.tail:
                    self.tail = current

                # Skip the node to be deleted
                current.next = current.next.next

                return

            # Move to next node
            current = current.next

        print("Value not found")

    # Function to remove all nodes from the list
    def clear(self):

        # Remove references to all nodes
        self.head = None
        self.tail = None

        print("Linked List Cleared")

    # Function to search for a value in the linked list
    def search(self, value):

        # Start from the head node
        current = self.head

        # Position counter (index)
        position = 0

        # Traverse the linked list
        while current is not None:

            # Check if current node contains the value
            if current.data == value:
                print(f"Value {value} found at position {position}")
                return True

            # Move to next node
            current = current.next

            # Increment position
            position += 1

        # Value not found after traversing entire list
        print(f"Value {value} not found")
        return False
    
    # Function to insert a new node at the beginning
    def insert_begin(self, data):

        # Create a new node
        new_node = Node(data)

        # If list is empty
        if self.head is None:

            # New node becomes both head and tail
            self.head = new_node
            self.tail = new_node
            return

        # New node points to current head
        new_node.next = self.head

        # Move head to new node
        self.head = new_node
```

```python
l1 = SinglyLinkedList()
```

```python
l1.insert_end(10)
l1.insert_end(20)
l1.insert_end(30)
l1.insert_end(40)
l1.insert_end(50)

print("Original List:")
l1.display()
```

```python
l1.insert_begin(8)
l1.display()
```

```python
print("\nSearching for 30:")
l1.search(30)
```

```python
print("\nDeleting 30:")
l1.delete(30)
l1.display()
```

```python
print("\nSize:", l1.size())
```

```python
print("\nClearing List:")
l1.clear()
l1.display()
```
