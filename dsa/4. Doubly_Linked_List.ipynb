```python
class Node:
    def __init__(self, data):

        # Store value
        self.data = data

        # Previous node pointer
        self.prev = None

        # Next node pointer
        self.next = None
```

```python
class DoublyLinkedList:

    def __init__(self):

        # First node of the list
        self.head = None


    # INSERT AT BEGINNING
    def insert_begin(self, data):

        # Create new node
        new_node = Node(data)

        # Empty list
        if self.head is None:

            self.head = new_node
            return

        # New node points to current head
        new_node.next = self.head

        # Current head points back to new node
        self.head.prev = new_node

        # Move head to new node
        self.head = new_node


    # INSERT AT END
    def insert_end(self, data):

        # Create new node
        new_node = Node(data)

        # Empty list
        if self.head is None:

            self.head = new_node
            return

        # Start from head
        temp = self.head

        # Reach last node
        while temp.next:
            temp = temp.next

        # Last node points to new node
        temp.next = new_node

        # New node points back to last node
        new_node.prev = temp

    # DISPLAY
    def display(self):

        temp = self.head

        while temp:

            print(temp.data, end=" <-> ")

            temp = temp.next

        print("None")


    # SEARCH
    def search(self, key):

        temp = self.head

        position = 0

        while temp:

            if temp.data == key:

                print(f"Key {key} found at position {position}")
                return True

            temp = temp.next
            position += 1

        print("Key not found")
        return False


    # Deletes first occurrence of key
    def delete(self, key):

        temp = self.head

        # Empty list
        if temp is None:

            print("List is empty")
            return

        # CASE 1 : Delete head node
        if temp.data == key:

            self.head = temp.next

            if self.head:
                self.head.prev = None

            return

        # Search for key
        while temp and temp.data != key:
            temp = temp.next

        # Key not found
        if temp is None:

            print("Key not found")
            return

        # Update next node
        if temp.next:
            temp.next.prev = temp.prev

        # Update previous node
        if temp.prev:
            temp.prev.next = temp.next


    # Counts total nodes
    def size(self):

        count = 0

        temp = self.head

        while temp:

            count += 1
            temp = temp.next

        return count

    # Deletes entire list
    def clear(self):

        self.head = None

        print("Doubly Linked List Cleared")
```

```python
dll = DoublyLinkedList()
```

```python
# Insert at End
dll.insert_end(10)
dll.insert_end(20)
dll.insert_end(30)

print("After Insert End:")
dll.display()
```

```python
# Insert at Beginning
dll.insert_begin(5)

print("\nAfter Insert Begin:")
dll.display()
```

```python
# Search
print("\nSearching for 20:")
dll.search(20)

print("\nSearching for 100:")
dll.search(100)
```

```python
# Delete
print("\nDeleting 20:")
dll.delete(20)

dll.display()
```

```python
# Size
print("\nSize of List:")
print(dll.size())
```

```python
# Clear
print("\nClearing List:")
dll.clear()

dll.display()
```
