```python
# Node class represents a single node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# Singly Linked List using only HEAD
class SinglyLinkedList:

    def __init__(self):
        # Pointer to first node
        self.head = None


    # Add node at the end
    def insert_end(self, data):

        # Create a new node
        new_node = Node(data)

        # If list is empty
        if self.head is None:
            self.head = new_node
            return

        # Start from head
        current = self.head

        # Move to the last node
        while current.next is not None:
            current = current.next

        # Attach new node at the end
        current.next = new_node


    # Count number of nodes
    def size(self):

        count = 0
        current = self.head

        while current is not None:
            count += 1
            current = current.next

        return count


    # Display linked list
    def display(self):

        current = self.head

        while current is not None:
            print(current.data, end=" -> ")
            current = current.next

        print("None")


    # Delete first occurrence of a value
    def delete(self, value):

        # Empty list
        if self.head is None:
            print("List is empty")
            return

        # If first node contains value
        if self.head.data == value:
            self.head = self.head.next
            return

        # Start traversal
        current = self.head

        while current.next is not None:

            # Check next node
            if current.next.data == value:

                # Skip the node
                current.next = current.next.next
                return

            current = current.next

        print("Value not found")


    # Remove all nodes
    def clear(self):

        self.head = None
        print("Linked List Cleared")

    # Search for a value in the linked list
    def search(self, value):

        # Start from the head node
        current = self.head

        # Position counter
        position = 0

        # Traverse the list
        while current is not None:

            # Check if value is found
            if current.data == value:
                print(f"Value {value} found at position {position}")
                return True

            # Move to next node
            current = current.next

            # Increment position
            position += 1

        # Value not found
        print(f"Value {value} not found")
        return False

    # Insert a node at the beginning of the list
    def insert_begin(self, data):

        # Create a new node
        new_node = Node(data)

        # New node points to current head
        new_node.next = self.head

        # Move head to new node
        self.head = new_node
```

```python
l1 = SinglyLinkedList()
```

```python
l1.insert_end(10)
l1.insert_end(20)
l1.insert_end(30)
l1.insert_end(40)
l1.insert_end(50)

print("Original List:")
l1.display()
```

```python
l1.insert_begin(8)
l1.display()
```

```python
print("\nSearching for 30:")
l1.search(30)
```

```python
print("\nDeleting 30:")
l1.delete(30)
l1.display()
```

```python
print("\nSize:", l1.size())
```

```python
print("\nClearing List:")
l1.clear()
l1.display()
```
