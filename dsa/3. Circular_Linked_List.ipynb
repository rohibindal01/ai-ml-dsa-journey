```python
# Node class represents a single node in the Circular Linked List
class Node:
    def __init__(self, data):
        self.data = data      # Store the value/data of the node
        self.next = None      # Pointer to the next node
```

```python
# Circular Linked List class
class CircularLinkedList:
    def __init__(self):
        self.head = None      # Initially, the list is empty


    # Insert a new node at the END of the list
    def insert_end(self, data):

        # Create a new node with the given data
        new_node = Node(data)

        # If list is empty
        if self.head is None:
            self.head = new_node          # Make new node the head
            new_node.next = self.head     # Point to itself (circular link)
            return

        # Start from head
        temp = self.head

        # Move until the last node
        # Last node's next points back to head
        while temp.next != self.head:
            temp = temp.next

        # Connect last node to new node
        temp.next = new_node

        # New node points back to head to maintain circularity
        new_node.next = self.head


    # Insert a new node at the BEGINNING of the list
    def insert_begin(self, data):

        # Create a new node
        new_node = Node(data)

        # If list is empty
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return

        # Find the last node
        temp = self.head

        while temp.next != self.head:
            temp = temp.next

        # New node should point to current head
        new_node.next = self.head

        # Last node should point to new node
        temp.next = new_node

        # Update head to new node
        self.head = new_node


    # Delete a node with a given value (key)
    def delete(self, key):

        # If list is empty
        if self.head is None:
            print("List is empty")
            return

        # Start from head
        curr = self.head

        # CASE 1: Head node contains the key
        if curr.data == key:

            # If only one node exists
            if self.head.next == self.head:
                self.head = None
                return

            # Find the last node
            temp = self.head

            while temp.next != self.head:
                temp = temp.next

            # Last node points to second node
            temp.next = self.head.next

            # Move head to next node
            self.head = self.head.next
            return

        # CASE 2: Node to delete is not the head
        prev = self.head
        curr = self.head.next

        # Traverse the list until we return to head
        while curr != self.head:

            # Key found
            if curr.data == key:
                prev.next = curr.next     # Skip the node
                return

            # Move both pointers forward
            prev = curr
            curr = curr.next

        # If key doesn't exist
        print("Key not found")


    # Display all nodes in the Circular Linked List
    def display(self):

        # If list is empty
        if self.head is None:
            print("List is empty")
            return

        # Start from head
        temp = self.head

        while True:

            # Print current node data
            print(temp.data, end=" -> ")

            # Move to next node
            temp = temp.next

            # Stop when we come back to head
            if temp == self.head:
                break

        print("(back to head)")
    
    # Search for a value in the Circular Linked List
    def search(self, key):

        # If list is empty
        if self.head is None:
            print("List is empty")
            return False

        # Start from head
        temp = self.head

        position = 0

        while True:

            # Check current node
            if temp.data == key:
                print(f"Key {key} found at position {position}")
                return True

            # Move to next node
            temp = temp.next

            position += 1

            # If we reach head again, stop
            if temp == self.head:
                break

        print("Key not found")
        return False
    
    # Count total nodes in Circular Linked List
    def size(self):

        # Empty list
        if self.head is None:
            return 0

        count = 0
        temp = self.head

        while True:

            count += 1

            temp = temp.next

            if temp == self.head:
                break

        return count
    
    # Remove all nodes from the list
    def clear(self):

        # Remove head reference
        self.head = None

        print("Circular Linked List Cleared")
```

```python
cll = CircularLinkedList()
```

```python
# Insert at End
cll.insert_end(10)
cll.insert_end(20)
cll.insert_end(30)

print("After Insert End:")
cll.display()
```

```python
# Insert at Beginning
cll.insert_begin(5)

print("\nAfter Insert Begin:")
cll.display()
```

```python
# Search
print("\nSearching for 20:")
cll.search(20)
```

```python
# Delete
print("\nDeleting 20:")
cll.delete(20)
cll.display()
```

```python
# Size
print("\nSize of Circular Linked List:")
print(cll.size())
```

```python
# Clear
print("\nClearing List:")
cll.clear()

cll.display()
```
